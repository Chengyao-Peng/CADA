```python
import xlrd
from xlutils.copy import copy


input_dir = '../data/raw/f2g/newTUcases_1_to_150.xlsx'
rb = xlrd.open_workbook(input_dir)
sheet1 = rb.sheets()[0]

wb = copy(rb)
sheet2 = wb.get_sheet(0)

for i in range(1, 1135):
    cell = sheet1.cell(i, 2).value
    num = len(cell)
    if num == 1:
        new_cell = 'HP:000000' + cell
    elif num == 2:
        new_cell = 'HP:00000' + cell
    elif num == 3:
        new_cell = 'HP:0000' + cell
    elif num == 4:
        new_cell = 'HP:000' + cell
    elif num == 5:
        new_cell = 'HP:00' + cell
    elif num == 6:
        new_cell = 'HP:0' + cell
    elif num == 7:
        new_cell = 'HP:' + cell
    
    sheet2.write(i,2,new_cell)
    wb.save('../data/raw/f2g/newTUcases_1_to_150.xlsx')





```
