```python
import pandas as pd

old = '/home/peng/PycharmProjects/CADA/data/raw/patients/clinvar/clinvar_submissions.tsv'
peter = '/home/peng/PycharmProjects/CADA/data/raw/patients/clinvar/clinvar_submissions_PK.tsv'

df_old = pd.read_csv(old, sep='\t')
df_peter = pd.read_csv(peter, sep='\t')[['patient_id', 'label']]
df_peter = df_peter.fillna('train')
df = pd.merge(df_old, df_peter, on='patient_id')

df['patient_id'] = 'Patient:' + df['patient_id']
```

```python
df[df.label == 'train'][['patient_id', 'omim', 'gene', 'hpos', 'submitter', 'from']].to_csv('/home/peng/PycharmProjects/CADA/data/raw/patients/clinvar/clinvar_submissions_training.tsv', sep='\t', index=None)
```

```python
df[df.label != 'train'].to_csv('/home/peng/PycharmProjects/CADA/data/raw/patients/clinvar/clinvar_submissions_testing.tsv', sep='\t', index=None)
```
