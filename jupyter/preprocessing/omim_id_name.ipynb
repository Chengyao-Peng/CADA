```python
import os
import pandas as pd
from CADA.paths import DATA_DIRECTORY
import pickle

input_dir = os.path.join(DATA_DIRECTORY, 'raw', 'mapping', 'mimTitles.txt')
out_id_name = os.path.join(DATA_DIRECTORY, 'processed', 'ids', 'omim_id_name.dict')

with open(input_dir) as infile:
    with open(out_id_name, 'wb') as outfile:
        df = pd.read_csv(input_dir, skiprows=5, header=None, sep='\t').applymap(str)
        dict_id_name = dict(zip("OMIM:" + df.iloc[:,1], df.iloc[:,2]))
        pickle.dump(dict_id_name, outfile)
```
