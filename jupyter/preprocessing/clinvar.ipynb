```python
import os
from CADA.paths import DATA_DIRECTORY
import pandas as pd

old = os.path.join(DATA_DIRECTORY, 'raw', 'patients', 'clinvar', 'clinvar_submissions.tsv')
peter = os.path.join(DATA_DIRECTORY, 'raw', 'patients', 'clinvar', 'clinvar_submissions_PK.tsv')

df_old = pd.read_csv(old, sep='\t')
df_peter = pd.read_csv(peter, sep='\t')[['patient_id', 'label']]
df_peter = df_peter.fillna('train')
df = pd.merge(df_old, df_peter, how='left', on='patient_id')
```

```python
df['label'].fillna("train",inplace = True)
df['patient_id'] = 'Patient:' + df['patient_id']
```

```python
clinvar_submissions_all = os.path.join(DATA_DIRECTORY, 'raw', 'patients', 'clinvar', 'clinvar_submissions_all.tsv')
df[['patient_id', 'omim', 'gene', 'hpos', 'submitter', 'from']].to_csv(clinvar_submissions_all, sep='\t', index=None)
```

```python
clinvar_submissions_training = os.path.join(DATA_DIRECTORY, 'raw', 'patients', 'clinvar', 'clinvar_submissions_training.tsv')
df[df.label == 'train'][['patient_id', 'omim', 'gene', 'hpos', 'submitter', 'from']].to_csv(clinvar_submissions_training, sep='\t', index=None)
```

```python
clinvar_submissions_testing = os.path.join(DATA_DIRECTORY, 'raw', 'patients', 'clinvar', 'clinvar_submissions_testing.tsv')
df[df.label != 'train'].to_csv(clinvar_submissions_testing, sep='\t', index=None)
```
