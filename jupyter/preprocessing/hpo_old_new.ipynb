```python
import os
import pickle
from pronto import Ontology
from CADA.paths import DATA_DIRECTORY
```

```python
in_dir = os.path.join(DATA_DIRECTORY, 'raw', 'hpo', 'hpo_hierarchical_information', 'hp.obo')
out_dir = os.path.join(DATA_DIRECTORY, 'processed', 'ids', 'hpo_old_new.dict')

with open(out_dir, 'wb') as outfile:
    old_new_hpo = {}
    hpo = Ontology(in_dir)
    for term in hpo.terms():
        id = term.id
        for alt_id in term.alternate_ids:
            old_new_hpo[alt_id] = id
    pickle.dump(old_new_hpo, outfile)
        
```
