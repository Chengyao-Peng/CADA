```python
from CADA.triples_gene_hpo import triples_gene_hpo
import pandas as pd
from CADA.paths import DATA_DIRECTORY

gene_hpo_triples = triples_gene_hpo()
```

```python
gene_numhpo = {}
for triple in gene_hpo_triples:
    gene = triple[2]
    hpo = triple[0]
    if gene in gene_numhpo:
        gene_numhpo[gene] += 1
    else:
        gene_numhpo[gene] = 1
gene_numhpo = pd.DataFrame.from_dict(gene_numhpo, orient='index', columns=['num_hpo_annotation']).sort_values(by='num_hpo_annotation', ascending=False)
gene_numhpo.to_csv('gene_no_hpo.txt', header=None, index=True, sep='\t')
```

```python
patients_df = pd.read_csv('patient.tsv',header=0, sep='\t')
gene_dict = dict()
for i in patients_df.groupby('gene_id').count().sort_values('patient_id', ascending=False).iterrows():
    gene_dict[i[0]] = i[1][0]

gene_numpatient = pd.DataFrame.from_dict(gene_dict, orient='index', columns=['num_patient'])
```
