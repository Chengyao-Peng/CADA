```python
import pandas as pd
import os
import shutil
```

```python
dir_name = 'phen2gene'
if os.path.exists(dir_name):
    print("Directory '%s' exist" %dir_name)
else:
    os.mkdir('phen2gene')
    print("Directory '%s' created" %dir_name)
```

```python
# read test patients
test_patients = pd.read_csv('patient_testing.tsv', sep = '\t', header=None)
for patient in test_patients.values.tolist():
    file_name = patient[0].split(':')[1]
    file_path = os.path.join(dir_name, file_name)
    hpos = '\n'.join(patient[3].split(','))
    with open(file_path, 'w') as f:
        f.write(hpos)
test_patients
```

```python
####run phen2gene manually and save result to phen2gene/out
evaluation = []
for patient in test_patients.values.tolist():
    patient_name = patient[0].split(':')[1]
    target_gene = patient[2].split(':')[1]
    result_path = os.path.join(dir_name, 'out', dir_name, patient_name, 'output_file.associated_gene_list')
    result_pd = pd.read_csv(result_path, sep = '\t', header = 0, dtype = str)
    try:
        rank = int(result_pd.loc[result_pd['ID'] == target_gene, 'Rank'].iloc[0])
    except IndexError:
        rank = 'NA'
    evaluation.append([patient[0], patient[2], patient[3], rank])
    
eval_df = pd.DataFrame(evaluation, columns = ['patient_id', 'gene_id', 'features', 'rank'])
eval_df
```

```python
# count top-n accuracy 
ranks = eval_df['rank'].tolist()
count1 = count5 = count10 = count50 = count100 = count1000 = 0
counts = len(ranks)
for rank in ranks:
    if isinstance(rank, int):
        if rank == 1:
            count1 += 1
        if rank <= 5:
            count5 += 1
        if rank <= 10:
            count10 += 1
        if rank <= 50:
            count50 += 1
        if rank <= 100:
            count100 += 1
        if rank <= 1000:
            count1000 += 1
top1 = round(count1/counts, 2)
top5 = round(count5/counts, 2)
top10 = round(count10/counts, 2)
top50 = round(count50/counts, 2)
top100 = round(count100/counts, 2)
top1000 = round(count1000/counts, 2)

print(f'top1: {top1}')
print(f'top5: {top5}')
print(f'top10: {top10}')
print(f'top50: {top50}')
print(f'top100: {top100}')
print(f'top1000: {top1000}')
```
