```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df_with_abs = pd.read_excel('../../../model/withoutgene_omim_with_patients/evaluation_with_absenthpo.xlsx')
df_without_abs = pd.read_excel('../../../model/withoutgene_omim_with_patients/evaluation.xlsx')[['results_withp','rank_withp']]
df = pd.concat([df_with_abs, df_without_abs], axis=1)
df = df[df.no_absfeatures > 0]
df['c']= np.negative(np.log(df.rank_withp_withabs/df.rank_withp))
df.to_excel('c.xlsx', index=None)

```

```python
import seaborn as sns; sns.set(color_codes=True)
ax = sns.regplot(x=df['No.features'], y=df['c'], data=df).get_figure()
ax.savefig("Nofeatures_c.png")
```

```python
ax = sns.regplot(x=df['No.patients'], y=df['c'], data=df).get_figure()
ax.savefig("Nopatients_c.png")
```
