```python
from gensim.models import Word2Vec
import os
import json
import pandas as pd
import numpy as np
from CADA.paths import DATA_DIRECTORY
import pickle
```

```python
model = Word2Vec.load('/home/peng/PycharmProjects/CADA/model/all_patients_disease_gene_doublefalse/with_patients.model')
```

```python
with open(os.path.join(DATA_DIRECTORY, 'processed', 'ids', 'gene_id_name.dict'), 'rb') as handle:
    gene_id_name = pickle.load(handle)

with open(os.path.join(DATA_DIRECTORY, 'processed', 'ids', 'hpo_id_name.dict'), 'rb') as handle:
    hpo_id_name = pickle.load(handle)
```

```python
input_dir = os.path.join(DATA_DIRECTORY, 'raw', 'patients', 'evaluation_from_jonathanz', 'cases.json')
results_tsv = []
results_excel = []

with open(input_dir) as infile:
    content = json.load(infile)
    for patient in content:
        features = []
        similar_gene_nodes = []
        similar_gene_nodes_name = []
        case = 'Patient:' + str(patient['case'])
        for feature in patient['hpo_num']:
            features.append('HP:' + str(feature.zfill(7)))
        if len(features) > 0:
            similar_nodes = model.most_similar(positive=features, topn=2000000)
            for node in similar_nodes:
                if node[0].startswith('Entrez'):
                    similar_gene_nodes.append(node[0] + '=' + str(node[1]))
                    similar_gene_nodes_name.append(gene_id_name[node[0]])
            results_tsv.append([case, ','.join(features), ','.join(similar_gene_nodes[:1000])])
            results_excel.append([case, ','.join([hpo_id_name[feature] for feature in features]), ','.join(similar_gene_nodes_name[:10])])
    df_tsv = pd.DataFrame(results_tsv, columns = ['patient_id', 'features', 'results'])
    df_tsv.to_csv('result.tsv', sep='\t', index=None)
    df_excel = pd.DataFrame(results_excel, columns = ['patient_id', 'features', 'results'])
    df_excel.to_excel('result.xlsx', index=None)
```
