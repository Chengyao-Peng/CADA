```python
from gensim.models import Word2Vec
import pandas as pd
import pronto

evaluation_patients = '/home/peng/PycharmProjects/featurematch/data/processed/patients/reform/patients_evaluation.tsv'
training_patients = '/home/peng/PycharmProjects/featurematch/data/processed/patients/reform/patients_training.tsv'
dict_patient_disease = {}

with open (training_patients, 'r') as infile:
    content = infile.read().splitlines()
    content = [x.split('\t') for x in content]
    
    for line in content:
        patient_id = line[0]
        disease = line[1]
        dict_patient_disease[patient_id] = disease 

model = Word2Vec.load('../../../model/mondo/70patient/70patients.model')


with open(evaluation_patients, 'r') as e_file:
    content = e_file.read().splitlines()
    content = [x.split('\t') for x in content]
    for line in content:
        patient_id = line[0]
        disease = line[1]
        features = line[2].split(',')
        gene = line[3]
        if line[4].split(',') == ['unknown']:
            absent_features = []
        else:
            absent_features = line[4].split(',')

        similar_nodes = model.most_similar(positive=features, negative=absent_features, topn=500)
        similar_nodes_features_filters = []
        for node in similar_nodes:
            if node[0].startswith('Patient'):
                similar_nodes_features_filters.append(node[0]+ '=' + dict_patient_disease[node[0]])
            elif node[0].startswith('Entrez'):
                similar_nodes_features_filters.append(node[0])
            elif node[0].startswith('MONDO'):
                similar_nodes_features_filters.append(node[0])
                
        print(patient_id + '\t' + disease + '\t' + gene + '\t' + str(len(features)) + '\t' + ','.join(similar_nodes_features_filters) + '\n')
```
