```python
import os
import pandas as pd
import networkx as nx
import collections
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick


indir = '/home/peng/PycharmProjects/featurematch/data/processed/final/with_patients/with_patients_mapped.triples'

df = pd.read_csv(indir, sep='\t', header=None).loc[:, [0, 2]]
edges = df.values.tolist()
tuple_edges = []

for edge in edges:
    tuple_edge = tuple(edge)
    tuple_edges.append(tuple_edge)

G = nx.Graph()
G.add_edges_from(tuple_edges)
    
```

```python
degs = {}
for n in G.nodes () :
    deg = G.degree (n)
    if deg not in degs :
        degs[deg] = 0
    degs[deg] += 1
items = sorted(degs.items())
print(items)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot([k for (k, v) in items],[v for (k, v) in items])
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel('Degree')
vals = ax.get_yticks()
ax.set_yticklabels(['{:,.2%}'.format(x) for x in vals])
plt.title ("Degree Distribution")
fig.savefig ("degree_distribution.png")
```
