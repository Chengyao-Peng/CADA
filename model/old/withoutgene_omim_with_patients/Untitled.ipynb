```python
from gensim.models import Word2Vec
import pandas as pd
from CADA.paths import DATA_DIRECTORY,MODEL_DIRECTORY
import pickle
import os

test = os.path.join(MODEL_DIRECTORY, 'with_patients_new', 'patient_testing.tsv')
evaluation_save = list()
evaluation_vis = list()
```

```python
model = Word2Vec.load('with_patients.model')
```

```python
with open(test, 'r') as t_file:
    content = t_file.read().splitlines()[1:]
    content = [x.split('\t') for x in content]
    for line in content:
        patient_id = line[0]
        gene_nodes = []
        gene_id = line[2]
        features = line[3].split(',')
        similar_nodes = model.most_similar(positive=features, topn=100000)
        for node in similar_nodes:
            if node[0].startswith('Entrez'):
                gene_nodes.append(node[0])
        rank = gene_nodes.index(gene_id) + 1
        print(patient_id + ' ' + rank)
```
