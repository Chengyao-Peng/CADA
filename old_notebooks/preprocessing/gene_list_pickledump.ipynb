```python
import os
from CADA.paths import DATA_DIRECTORY
import pandas as pd
import pickle
```

```python
in_file = os.path.join(DATA_DIRECTORY, 'raw', 'hpo', 'gene-hpo', 'ALL_SOURCES_ALL_FREQUENCIES_genes_to_phenotype.txt')
gene_list_file = os.path.join(DATA_DIRECTORY, 'processed', 'hpo', 'gene-hpo', 'genes.list')

with open(in_file) as infile:
    with open(gene_list_file, 'wb') as handle:
        genes = []
        df = pd.read_csv(infile, skiprows=1, header=None, sep='\t').applymap(str)
        for row in df.values.tolist():
            gene = row[0].strip()
            if gene not in genes:
                genes.append('Entrez:' + gene)
        pickle.dump(genes, handle)
```
