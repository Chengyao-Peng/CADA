```python
import os
import pandas as pd
from CADA.paths import DATA_DIRECTORY
import pickle

input_dir = os.path.join(DATA_DIRECTORY, 'raw', 'mapping', 'gene_result.txt')
out_id_name = os.path.join(DATA_DIRECTORY, 'processed', 'ids', 'gene_id_name.dict')
out_name_id = os.path.join(DATA_DIRECTORY, 'processed', 'ids', 'gene_name_id.dict')


with open(input_dir) as infile:
    with open(out_id_name, 'wb') as outfile_id_name:
        df1 = pd.read_csv(input_dir, skiprows=1, header=None, sep='\t').applymap(str)
        dict_id_symbol = dict(zip("Entrez:" + df1.iloc[:,2], df1.iloc[:,5]))
        pickle.dump(dict_id_symbol, outfile_id_name)

with open(input_dir) as infile:
    with open(out_name_id, 'wb') as outfile_name_id:
        df1 = pd.read_csv(input_dir, skiprows=1, header=None, sep='\t').applymap(str)
        dict_symbol_id = dict(zip(df1.iloc[:,5], 'Entrez:' + df1.iloc[:,2]))
        pickle.dump(dict_symbol_id, outfile_name_id)
```
